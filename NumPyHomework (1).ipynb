### 12.* 
Make a smoother image by applying a filter 
- hint: use a convolution

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('noisy.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
import imageio

im = imageio.imread('noisy.png')

im = np.asarray(im)

im.T
```

```python


kernel=np.array([[[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]],[[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]],[[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]]])
kernel=kernel/273
```

```python
n=800
m=5
```

```python
matrix2=np.zeros((n-m+1,n-m+1))
print(matrix2)
for i in range(0,n-m+1):
    for t in range(0,n-m+1):
        matrix2[i][t]=np.sum(np.multiply(im.T[:,i:(i+m),t:(t+m)],kernel))
        
        
        
    
```

```python
matrix2
```

```python
plt.imshow(matrix2.T,cmap='gray')
```

```python
imageio.imwrite('astronaut-gray.jpg', matrix2.T)
```
